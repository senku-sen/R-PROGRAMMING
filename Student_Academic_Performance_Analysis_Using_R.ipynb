 # STUDENT PERFORMANCE DATA ANALYSIS

```r
# Load Dataset

studentsPerformanceData <- read.csv("StudentsPerformance.csv")

# Display the current data
studentsPerformanceData
```

# PART 1: DATA EXPLORATION

```r
# Display First 10 Rows
head(studentsPerformanceData, 10)
```

```r
# Check Structure of Dataset
str(studentsPerformanceData)
```

```r
# Summary Statistics
summary(studentsPerformanceData)
```

```r
# Identify Data Types

sapply(studentsPerformanceData, class)
```

```r
# Count Number of Students per Gender
table(studentsPerformanceData$Gender)
```

# PART 2 DESCRIPTIVE STATISTICS

```r
# Mean Study Hours
mean(studentsPerformanceData$Study_Hours)
```

```r
# Median Exam Score
median(studentsPerformanceData$Exam_Score)
```

```r
# Standard Deviation of Quiz Scores
sd(studentsPerformanceData$Quiz_Score)
```

```r
# Minimum & Maximum Attendance Rate

min(studentsPerformanceData$Attendance_Rate)
max(studentsPerformanceData$Attendance_Rate)
```

```r
# Range of Exam Scores
range(studentsPerformanceData$Exam_Score)
```

# PART 3: GROUPED ANALYSIS

```r
# Mean Exam Score per Gender
tapply(studentsPerformanceData$Exam_Score, studentsPerformanceData$Gender, mean)
```

```r
# Mean Study Hours per Final Grade
tapply(studentsPerformanceData$Study_Hours, studentsPerformanceData$Final_Grade, mean)
```

```r
# Number of Students per Grade
table(studentsPerformanceData$Final_Grade)
```

```r
# Average Attendance Rate by Gender
tapply(studentsPerformanceData$Attendance_Rate, studentsPerformanceData$Gender, mean)
```

```r
# Grade Category with Highest Average Study Hour
mean_hours <- tapply(studentsPerformanceData$Study_Hours, studentsPerformanceData$Final_Grade, mean)

mean_hours
names(mean_hours)[which.max(mean_hours)]
```

# PART 4: DISTRIBUTION

```r
# Create Quiz Score Categories
studentsPerformanceData$Quiz_Category <- cut(
  studentsPerformanceData$Quiz_Score,
  breaks = c(60, 70, 80, 90, 100),
  labels = c("60-70", "71-80", "81-90", "91-100"),
  include.lowest = TRUE
)
```

# PART 4: DATA VISUALIZATION

```r
# Histogram of Exam Scores
hist(studentsPerformanceData$Exam_Score,
     main = "Histogram of Exam Scores",
     xlab = "Exam Score",
     ylab = "Frequency",
     col = "lightblue",
     border = "black",
     breaks = 10)
```

```r
# Boxplot of Exam Score by Gender
boxplot(Exam_Score ~ Gender,
        data = studentsPerformanceData,
        main = "Boxplot of Exam Score by Gender",
        xlab = "Gender",
        ylab = "Exam Score",
        col = c("pink", "lightgreen"),
        border = "black")
```

```r
# Scatterplot: Study Hours vs Exam Score
plot(studentsPerformanceData$Study_Hours, studentsPerformanceData$Exam_Score,
     main = "Study Hours vs Exam Score",
     xlab = "Study Hours per Week",
     ylab = "Exam Score",
     col = "darkblue",
     pch = 19)

# Add regression line
abline(lm(studentsPerformanceData$Exam_Score ~ studentsPerformanceData$Study_Hours),
       col = "red",
       lwd = 2)
```
